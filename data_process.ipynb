```python
def fill_with_zeros(value):
    if pd.isna(value) or value == '':
        return [0, 0, 0, 0, 0]
    else:
        try:
            # 如果已经是列表格式，直接返回
            if isinstance(value, list):
                return value
            # 如果是字符串格式，尝试将其转换为列表
            return eval(value)
        except (SyntaxError, TypeError):
            return [0, 0, 0, 0, 0]

def fill_with_zero_or_default(value, default=0):
    if pd.isna(value) or value == '':
        return default
    return value
```

```python
# 应用填充函数到 Neighbor1 到 Neighbor4 列
neighbor_columns = ['Neighbor1', 'Neighbor2', 'Neighbor3', 'Neighbor4']
for col in neighbor_columns:
    df_cleaned[col] = df_cleaned[col].apply(fill_with_zeros)
```

```python
# 定义转换函数
def convert_to_numeric_list(value):
    try:
        # 尝试将每个元素转换为浮点数
        return [float(x) for x in value]
    except (ValueError, TypeError):
        # 如果转换失败，返回全零列表
        return [0, 0, 0, 0, 0]
```

```python

df=data_cleaned_csv
# 定义一个函数，用于将字符串列表转换为数值列表
def str_list_to_num_list(str_list):
    # 去掉中括号并按逗号分割字符串，然后将每个元素转换为整数
    return [int(item.strip(" '")) for item in str_list.strip('[]').split(', ') if item.strip(" '")]

# 应用函数到'Serv'和'Neighbor'相关的列
columns_to_convert = ['Neighbor1', 'Neighbor2', 'Neighbor3', 'Neighbor4']
for col in columns_to_convert:
    df[col] = df[col].apply(str_list_to_num_list)

# 现在DataFrame中的'Serv'和'Neighbor'列已经转换为数值列表
print(df)
df.info()
```

```python
df.to_csv('cleaned_output_data_number_list.csv', index=False)
```

```python
import numpy as np
# 定义要展开的列
columns_to_expand = ['Serv', 'Neighbor1','Neighbor2','Neighbor3','Neighbor4']

# 计算每列的最大长度
max_lengths = {col: df[col].apply(len).max() for col in columns_to_expand}

# 创建一个新的 DataFrame 来保存展开后的数据
expanded_data = {}

# 处理需要展开的列
for col, max_length in max_lengths.items():
    # 使用 apply 和 lambda 函数来展开列表，并填充缺失值
    expanded_cols = df[col].apply(lambda x: x + [np.nan] * (max_length - len(x)))
    # 将展开后的列转置为多个新列
    for i in range(max_length):
        expanded_data[f'{col}_{i+1}'] = expanded_cols.apply(lambda x: x[i] if i < len(x) else np.nan)

# 保留原始 DataFrame 中不需要展开的列
for col in df.columns:
    if col not in columns_to_expand:
        expanded_data[col] = df[col]

# 创建新的 DataFrame
expanded_df = pd.DataFrame(expanded_data)

# 转换为 NumPy 数组
numpy_array = expanded_df.to_numpy()

# 打印结果
print("Expanded DataFrame:")
print(expanded_df)
print("\nNumPy Array:")
print(numpy_array)
```
